```python
import tensorflow.keras.models as models
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers

# Wczytaj model z pliku H5
model = models.load_model('moj_model.h5')
from google.colab import drive
drive.mount('/content/drive')
```

```python

img_width = 180
img_height =180

data_train_path = '/content/drive/My Drive/uczeniemaszynowe/Dataset_podzielony/train'

data_train = tf.keras.utils.image_dataset_from_directory(
    data_train_path,
    shuffle=True,
    image_size=(img_width, img_height),
    batch_size=32,
    validation_split=False)

data_cat = data_train.class_names

image = 'stadion.jpg'
image = tf.keras.utils.load_img(image, target_size=(img_height,img_width))
img_arr = tf.keras.utils.array_to_img(image)
img_bat=tf.expand_dims(img_arr,0)
predict = model.predict(img_bat)
score = tf.nn.softmax(predict)
print('In image is {} with accuracy of {:0.2f}'.format(data_cat[np.argmax(score)],np.max(score)*100))
plt.imshow(img_arr)
plt.axis('off')
plt.show()
```
